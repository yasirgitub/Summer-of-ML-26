### Which pokemon has the highest combined value of Attack and Sp. Atk ?

```python
best_att=(file["Attack"] + file["Sp. Atk"]).max()
result = file[(file["Attack"] + file["Sp. Atk"]) == best_att]
print(result.index)
```

### Which type of legendary pokemons are the most common?

```python
legendary = file[file["Legendary"] == True]

type1 = legendary["Type 1"].value_counts()
type2 = legendary["Type 2"].value_counts()

counts = type1.add(type2, fill_value=0)

print("Most common type:", counts.idxmax())
print("Count:", counts.max())
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
nonleng_gen3 = file[(file["Legendary"]==0)& (file["Generation"]==3)]
leng_gen6 = file[(file["Legendary"]==1)& (file["Generation"]==6)]
weakest_hp = leng_gen6['HP'].min()
count = (nonleng_gen3['HP'] > weakest_hp).sum()
print(count)
```
