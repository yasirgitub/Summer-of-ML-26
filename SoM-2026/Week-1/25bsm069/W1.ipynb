### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('../W1_Pokemon.csv')


type_totals = df.groupby('Type 1')['Total'].mean().reset_index()


type_totals = type_totals.sort_values('Total', ascending=False)


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
ax = sns.barplot(data=type_totals, x='Type 1', y='Total', palette='Set2')


plt.title('Average Total Stats by Primary Pokémon Type')
plt.xlabel('Primary Type')
plt.ylabel('Average Total Stat')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('../W1_Pokemon.csv')


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=df,
    x='Speed',
    y='Total',
    hue='Legendary',
    palette='Set1',
    alpha=0.7
)


plt.title('Pokémon: Total Stats vs. Speed')
plt.xlabel('Speed')
plt.ylabel('Total Stats')
plt.legend(title='Legendary')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('../W1_Pokemon.csv')

# Calculate Generation-wise average for Total stats
gen_avg = df.groupby('Generation')['Total'].mean().reset_index()


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.lineplot(data=gen_avg, x='Generation', y='Total', marker='o', color='mediumseagreen')

plt.title('Average Total Stats by Pokémon Generation')
plt.xlabel('Generation')
plt.ylabel('Average Total Stats')
plt.xticks(gen_avg['Generation'])
plt.tight_layout()
plt.show()

```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?


```python

import pandas as pd

df = pd.read_csv("../W1_Pokemon.csv")

bhangbharosa = (
    (df["Generation"] == 4)
    & (df["Legendary"] == False)
    & (df["Type 1"] == "Ghost")
)

secondfastest_non_legendary_ghost_type_pokemon = (
    df[bhangbharosa]
    .sort_values("Speed", ascending=False)
    .iloc[1]["Name"]
)
secondfastest_non_legendary_ghost_type_pokemon
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
import pandas as pd

df = pd.read_csv("../W1_Pokemon.csv")
char_def = df.loc[df['Name'] == 'Charizard', 'Defense'].iloc[0]
char_att = df.loc[df['Name'] == 'Charizard', 'Attack'].iloc[0]

count = df[
    (df['Legendary'] == False)
    & (df['Defense'] > char_def)
    & (df['Attack'] < char_att)
].shape[0]
count



```

### Which type of legendary pokemons are the most common?

```python
#your code here
import pandas as pd

df = pd.read_csv("../W1_Pokemon.csv")

# 1. Filter out only the legendary Pokémon
legendaries = df[df['Legendary'] == True]

# 2. Combine the Type 1 and Type 2 columns into a single series
all_legendary_types = pd.concat([legendaries['Type 1'], legendaries['Type 2']])

# 3. Drop missing values (since some Pokémon don't have a Type 2) and count occurrences
type_counts = all_legendary_types.dropna().value_counts()

most_common_type = type_counts.idxmax()
most_common_type












```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
import pandas as pd

df = pd.read_csv("../W1_Pokemon.csv")

# 1. Find the HP of the weakest Gen 6 Legendary Pokémon
gen6_legendaries = df[(df['Generation'] == 6) & (df['Legendary'] == True)]
weakest_gen6_hp = gen6_legendaries['HP'].min()

# 2. Filter for Gen 3, Non-Legendary Pokémon with higher HP than that value
matching_pokemon = df[
    (df['Generation'] == 3) & 
    (df['Legendary'] == False) & 
    (df['HP'] > weakest_gen6_hp)
]

# 3. Print the final numeric count
print(matching_pokemon.shape[0])

```
