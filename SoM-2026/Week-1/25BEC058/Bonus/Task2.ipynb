<h2>Task 2 : Visualization of results using Matplotlib library</h2>

<span style="font-size:24px;">**(Q1)** Display the data of top 5 east asian countries based on 'Generosity'. </span>


```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df1 = pd.read_csv("world-happiness-report-2021.csv")
top5 = df1[df1["Regional indicator"] == 'East Asia'].sort_values("Generosity",ascending=False).head(5)
top5
```

<span style="font-size:24px;">**(Q2)** Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity).</span>


```python
sns.lineplot(x="Country name",y="Ladder score",data = top5,marker="o",color="green")
plt.title("Ladder score vs Top 5 East Asian Countries(Based on Generoisty)")
plt.xticks(rotation=60)
plt.show()
```

<span style="font-size:24px;">**(Q3)** Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'.</span>


```python
df_2021 = df1[(df1["Country name"] == 'India') | (df1["Country name"] == 'China') | (df1["Country name"] == 'Nepal') | (df1["Country name"] == 'Bangladesh') | (df1["Country name"] == 'Pakistan') | (df1["Country name"] == 'Myanmar') | (df1["Country name"] == 'Afghanistan')]
sns.scatterplot(x="Country name",y="Logged GDP per capita",data =df_2021,marker="D",color="tomato")
plt.title("Country vs their Logged GDP per captia")
plt.show()
```
