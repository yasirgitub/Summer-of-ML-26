### Task 2: Visualization of results using Matplotlib library


1.Display the data of top 5 east asian countries based on 'Generosity'


```python
data = df1[df1['Regional indicator']=='East Asia'].sort_values(by=['Generosity']).head()
data
```

2.Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)


```python
import matplotlib.pyplot as plt
plt.plot(data['Generosity'].values, data['Ladder score'].values,label=['ladder score'])
plt.legend()
```

3.Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'


```python

df_2021= df1.set_index('Country name')
df_2021 = df_2021.loc[['China' , 'Nepal' ,'Bangladesh' ,'Pakistan' , 'Myanmar' , 'India' ,'Afghanistan']]


plt.scatter(df_2021.index , df_2021['Logged GDP per capita'],label= 'gdp per capita')
plt.legend()
```
