### 7) Which was the most-ordered item?

```python
# your code here
most_ordered_item = Chipo["item_name"].value_counts().index[0]
print(most_ordered_item)
```

### 8) For the most-ordered item, how many items were ordered?

```python
# your code here
print(Chipo["item_name"].value_counts()[most_ordered_item])
```

### 9) What was the most ordered item in the choice_description column?

```python
# your code here
most_ordered_item_cd = Chipo["choice_description"].value_counts().index[0]
print(most_ordered_item_cd)
```

### 10) How many items were ordered in total?

```python
# your code here
print(Chipo["quantity"].sum())
```

### 11) Turn the item price into a float

```python
# your code here
Chipo["item_price"] = (
    Chipo["item_price"].str.replace("$", "", regex=False).astype(float)
)

Chipo.dtypes
```

### 12) How much was the revenue for the period in the dataset?
#### Revenue will be the sum of quantity * item_price!

```python
# your code here
Chipo["revenue"] = Chipo["quantity"] * Chipo["item_price"]
print(Chipo["revenue"].sum())
```

### 13) How many orders were made in the period?

```python
# your code here
len(Chipo["order_id"].unique())
```

### 14) What is the average revenue amount per order?

```python
# your code here
print(np.round(np.divide(Chipo["revenue"].sum(), len(Chipo["order_id"].unique())), 2))
```

### 16) Show a histogram with number of orders with bill on the x-axis.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,12))

data = Chipo.groupby("order_id")["revenue"].sum()
plt.hist(data, edgecolor='black', color='skyblue',histtype='stepfilled',rwidth=0.4)
plt.show()

```
