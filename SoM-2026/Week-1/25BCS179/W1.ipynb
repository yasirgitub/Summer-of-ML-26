### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
ghost_True = (df['Type 1'] == 'Ghost') | (df['Type 2'] == 'Ghost')
gen_True = (df['Generation'] == 4)
speedMax = df.loc[(ghost_True & gen_True),'Speed'].max()
secondMax = 0
for speed in df.loc[(ghost_True & gen_True),'Speed']:
    if secondMax < speed < speedMax:
        secondMax = speed
df[(df['Speed'] == secondMax) & gen_True & ghost_True]
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
chaAttack = df.loc['Charizard','Attack']
chaDefense = df.loc['Charizard','Defense']
print(((df['Attack'] < chaAttack) & (df['Defense'] > chaDefense) & (df['Legendary'] == False)).sum())
```
