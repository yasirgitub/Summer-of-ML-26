### 16) Which profession has youngest people(apart from student)? Show the trend of mean age of people with profession as the seperating factor.

```python
#your code here
not_stud=user[user["occupation"]!="student"]
print(not_stud.groupby("occupation")["age"].mean().idxmin())
import matplotlib.pyplot as plt
meanage = not_stud.groupby("occupation")["age"].mean()

meanage.sort_values().plot(kind='barh')
plt.xlabel('Mean Age')
plt.ylabel('Profession')
plt.title('Mean Age by Profession (excluding students)')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```
