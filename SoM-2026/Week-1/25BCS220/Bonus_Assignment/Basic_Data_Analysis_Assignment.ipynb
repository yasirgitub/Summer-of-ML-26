### 11) Summarize the DataFrame.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(user['age'], kde=True)
plt.title("Age Distribution")
plt.show()

sns.countplot(user,x='gender')
plt.show()

plt.figure(figsize=(12,5))
sns.countplot(user,x='occupation')
plt.xticks(rotation=90)
plt.show()
```

### 12) Summarize only the occupation column

```python
plt.figure(figsize=(15,6))
sns.countplot(user,x='occupation', hue='gender')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15,6))
sns.boxplot(user,x='occupation', y='age')
plt.xticks(rotation=90)
plt.show()
```

### 14) What is the age with least occurrence?

```python
for i in user['age'].value_counts()[user['age'].value_counts() == user['age'].value_counts().min()].index:
    print(i)
```

These are the ages with least occurrence(I mean least non-zero occurrence :) ), These ages only occur 1 time.

### 15) What is the zip code with most occurrence?

```python
for i in user['zip_code'].value_counts()[user['zip_code'].value_counts()==user['zip_code'].value_counts().max()].index:
    print(i)
```

This is singular zip code occurs maximum number of times, i could have used mode, that would be better but I learnt new thing so I used this.

### 16) Which profession has youngest people(apart from student)? Show the trend of mean age of people with profession as the seperating factor.

```python
non_student = user[user['occupation'] != 'student']
p = non_student.groupby('occupation').age.median()[non_student.groupby('occupation').age.median() == non_student.groupby('occupation').age.median().min()].index
for i in p:
    print(i)
```
