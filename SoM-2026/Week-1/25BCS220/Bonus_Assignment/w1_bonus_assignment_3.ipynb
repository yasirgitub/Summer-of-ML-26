```python
t5ea = df1[df1['Regional indicator']=='Southeast Asia'].sort_values('Generosity', ascending=False).head()
t5ea
```

```python
plt.figure(figsize=(8,5))
plt.plot(t5ea['Country name'], t5ea['Ladder score'], marker='.')
plt.title('Ladder Score of Top 5 East Asian Countries by Generosity')
plt.xlabel('Country')
plt.ylabel('Ladder Score')
plt.grid(True)
plt.show()
```

```python
sea = ['India','China','Bangladesh', 'Nepal', 'Pakistan', 'Myanmar', 'Afganistan']
df_2021 = df1[df1['Country name'].isin(sea)]
plt.figure(figsize=(10,5))
plt.scatter(df_2021['Country name'], df_2021['Logged GDP per capita'])
plt.title('Country vs Logged GDP per Capita')
plt.xlabel('Country')
plt.ylabel('Logged GDP per capita')
plt.show()
```
